```python
import pandas as pd

data=pd.read_csv('northern-ireland-local-authority-collected-municipal-waste-management-statistics-time-series-1.csv')
data
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
import matplotlib.pyplot as plt
from sklearn import metrics
import joblib
```

```python
outputObj=data.iloc[:,7]
output=outputObj.tolist()
len(output)
```

```python
inputObj=data.iloc[:,35]
inputs=inputObj.tolist()
len(inputs)
```

```python
int_x=[]
int_y=[]

for i in range(1,1128):
    intX=int(inputs[i].replace(',', ''))
    intY=int(output[i].replace(',', ''))
    
    int_x.append(intX)
    int_y.append(intY)
    
int_x
```

```python
x = np.array(int_x).reshape((-1, 1))
y = np.array(int_y)
```

```python
model = LinearRegression()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

```python
print("Training feature size :",len(x_train))
print("Test feature size :",len(x_test))
print("Training output size :",len(y_train))
print("Test output size :",len(y_test))
```

```python
linear_model=model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
g=plt.plot(y_test - y_pred,marker='o',linestyle='')
```

```python
g=plt.scatter(y_test, y_pred)
g.axes.set_yscale('log')
g.axes.set_xscale('log')
g.axes.set_xlabel('True Values ')
g.axes.set_ylabel('Predictions ')
g.axes.axis('equal')
g.axes.axis('square')
```

```python
mean_squared_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
